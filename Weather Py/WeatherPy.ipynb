```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
import random
from citipy import citipy
import requests
import json

# Google API Key
from config import gkey
from config import weather_api_key


# weather url
url = "http://api.openweathermap.org/data/2.5/weather?"
```

## Generate Cities List

```python
# create random floats for long (-180 to +180) and lat (-90 to +90), use these in a loop to search citipy
lats =[]
longs = []

while len(lats) <2000:
    lat_random = random.uniform(-89.99, 99.99)
    lats.append(lat_random)

while len(longs)<2000:
    long_random = random.uniform(-179.99, 179.99)
    longs.append(long_random)
print('lats = '+str(len(lats)))
print('longs = '+str(len(longs)))

```

```python
#make lat longs lists array and then zip for looping by citypy
lat_array = np.array(lats)
lng_array = np.array(longs)
lat_long_array = zip(lat_array, lng_array)
```

```python
city_names=[]

# Use for loop to loop through lats and lgns with citypy function to return city names
for lat_lng in lat_long_array:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1])
    cityname = city.city_name

    if cityname not in city_names:
        city_names.append(cityname)

#print('Number of city names = '+str(len(city_names)))
print (len(city_names))
  
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Build query URL and loop through api call to retrieve 500+ records

city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

#counter
record_counter = 0

for city in city_names:
    try:
        query_url= "http://api.openweathermap.org/data/2.5/weather?" + "appid=" + weather_api_key + "&q=" + city + "&units=metric"
        response = requests.get(query_url).json()
        #print(response)
        city_name.append(response['name'])
        max_temp.append(response["main"]["temp_max"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        wind_speed.append(response['wind']['speed'])
        record_counter = record_counter + 1
        print (f"Appended record {record_counter} of set 1 for | {city}")
        # Wait a second in loop to not over exceed rate limit of API
        time.sleep(1.01)
    except:
        print("City not in data, processing next city...")
        

```

## Linear Regression

```python
#Northern and Southern Hemisphere data frames 
nrth_cities =cities_df.loc[cities_df["Lat"]>=0.1, :]
print('The number of cities in the Northern Hemisphere = '+str(len(nrth_cities)))


sth_cities =cities_df.loc[cities_df["Lat"]<=-0.1,:] 
print('The number of cities in the Southern Hemisphere = '+str(len(sth_cities)))
```

```python
#Lists for Northern 
nrth_lat_list = nrth_cities['Lat'].tolist()
nrth_temp_list = nrth_cities['Max Temp'].tolist()
nrth_humid_list = nrth_cities['Humidity'].tolist()
nrth_cloud_list = nrth_cities['Cloudiness'].tolist()
nrth_wind_list = nrth_cities['Wind Speed'].tolist()

#Lists for Southern
sth_lat_list = sth_cities['Lat'].tolist()
sth_temp_list = sth_cities['Max Temp'].tolist()
sth_humid_list = sth_cities['Humidity'].tolist()
sth_cloud_list = sth_cities['Cloudiness'].tolist()
sth_wind_list = sth_cities['Wind Speed'].tolist()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Regression nrth - temp vs lat
nrt_temp_lat_correlation = st.pearsonr(nrth_lat_list, nrth_temp_list)
#Northern Hemisphere - Temperature (C) vs. Latitude plot
plt.figure(figsize=(8,5))
plt.grid(True)
# converting lat result list to array
nrth_lat_arr = np.array(nrth_lat_list)
#create plot
x_values = nrth_lat_arr
y_values = nrth_temp_list
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Northern Latitude")
plt.ylabel("Max Temperature (C)")
r_squared = round((rvalue**2),2)
print(f"The correlation between Northern Hemisphere Latitute and Max Temperature (C) is: {round(nrt_temp_lat_correlation[0],2)}")
print(f"with an r-squared value of: {r_squared}")
plt.title("Northern Hemisphere - Temperature (C) vs. Latitude")
plt.annotate(line_eq,(50,25),fontsize=12,color="red")
plt.savefig('../Images/Nrth_Lat_Temp_correlation.png')
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Southern Hemisphere - Temp vs. Lat regression
sth_temp_lat_correlation = st.pearsonr(sth_lat_list, sth_temp_list)
#Southern Hemisphere - Temperature (C) vs. Latitude plot
plt.figure(figsize=(8,5))
plt.grid(True)
# converting lat result list to array
sth_lat_arr = np.array(sth_lat_list)
#create plot
x_values = sth_lat_arr
y_values = sth_temp_list
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Southern Latitude")
plt.ylabel("Max Temperature (C)")
r_squared = round((rvalue**2),2)
print(f"The correlation between Southern Hemisphere Latitute and Max Temperature (C) is: {round(sth_temp_lat_correlation[0],2)}")
print(f"with an r-squared value of: {r_squared}")
plt.title("Southern Hemisphere - Temperature (C) vs. Latitude")
plt.annotate(line_eq,(43,16),fontsize=12,color="red")
plt.savefig('../Images/Sth_Lat_Temp_correlation.png')
plt.show()
```

The above Pearsons r linear regressions provides statistical analysis of the relationship between maximum temperature (C) (for the day on which the sample was extracted, Oct 2020) and latitude for cities in each hemisphere.  As expected, there is a strong relationship between maximum temperature (C) and latitude, with correlations calculated, nothern = 0.84 and southern =0.55.  The relationship is very strong in the northern hemisphere sample as the r squared value shows that 71% of the maximum temperatures of city's can be predicted by latitudinal position.  This raltionship is present in the southern hemisphere sample too, however not as strong, with only 30% of maximum temperature predictable by a city's southern latitude. The smaller sample size (n=202) of the southern hemisphere cities, compared to northern cities (n=433) might be in part a cause of this weaker correlation result. Latitude is however, a good predicter of a city's maximum temperature.
